```python

def clean_tweet(column_name):
    name=set()
    for name_string in tweets1[column_name]:
        name.update(name_string.split(', '))
    name=sorted(name)
    return name

genres=clean_tweet('genres')
#Add one column for every genre in the dataframe
for genres in genres:
    tweets1["genres:"+genres] = [int(genres in g.split(', ')) for g in tweets1.genres]
    
keywords=clean_tweet('keywords')
#Add one column for every genre in the dataframe
for keywords in keywords:
    tweets1["keywords:"+keywords] = [int(keywords in g.split(', ')) for g in tweets1.keywords]
```

```python
def plot_column(column_name,n_elem_display=0):
    name=clean(column_name)
    dummy_name=pd.DataFrame()
    for n in name:
        dummy_name[n]=[int(n in nm.split(', ')) for nm in tweets1[column_name]] 
    
    namelist=[n for n in name]
    nlt=dummy_name[namelist].sum()
    nlt=nlt.sort_values(axis=0,ascending=False)
    if n_elem_display !=0:
     return nlt[:n_elem_display].plot(kind = "bar",figsize=(50,50))
    else:
     return nlt[:].plot(kind = "bar",figsize=(10,5))
```
